```python
import pandas as pd
import plotly.express as px
import datetime
from pytimeparse.timeparse import timeparse
from datetime import timedelta
import json


with open(f'./rostering.json', 'r', encoding='utf-8') as f:
        rostering = json.load(f)
        


format = '%d.%m.%y %H:%M'
shifts = map(
        lambda x: dict(Task = str(x['employeeId']),
                       Start = datetime.datetime.strptime(f"{x['shiftDate']} {x['shiftTimeStart']}", format),
                       Finish = datetime.datetime.strptime(f"{x['shiftDate']} {x['shiftTimeStart']}", format) + datetime.timedelta(seconds=32400),
                       Resource = x['shiftId'])
        , rostering['campainSchedule']
)

df_shifts = pd.DataFrame(shifts)
df_shifts.head()

```

```python
fig = px.timeline(df_shifts, x_start="Start", x_end="Finish", y="Task", color="Resource", height = 10000)
fig.update_yaxes(autorange="reversed")
# fig.update_xaxes(rangeslider_visible=True)
fig.show()

```
